## Grafiklere Argüman verme

```python
x = np.linspace(0,10,100) # 0 ile 10 arasında 100 eşit sayı üretti.
y = np.sin(x)
plt.figure(figsize = (10,5)) # Burada size ayarlandı.
plt.plot(x,y,"m4") # Burada m4 argumanlardan seçildi.
plt.show()
```

```python
plt.figure(figsize = (10,5)) # Burada size ayarlandı.
plt.plot(x,y,"y^") # Burada argumanlardan seçildi.
plt.show()
```

# Axis Name, Title, 

```python
plt.figure(figsize=(8, 5)) # axes in boyutu
plt.plot(x, y, "r*") # Grafik oluşturma ve çizgi belirleme 
plt.xlabel("x Axis") # x ekseni isim yazma
plt.ylabel("Y Axis") # y ekseni isim yazma
plt.title(label = "AXES") # Axes başlık atma
plt.show() #Gösterme
```

```python
plt.figure(figsize=(8, 5))
plt.plot(x, y, "b-.")
plt.xlabel("x Axis")
plt.ylabel("Y Axis")
plt.title(label = "AXES")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(x, y, "g:")
plt.xlabel("x Axis")
plt.ylabel("Y Axis")
plt.title(label = "AXES")
plt.show()
```

```python
age = np.arange(25, 44, 2.5)
salary_1 = np.arange(40000, 76000, 5000)
salary_2 = [45372, 46852, 69401, 56790, 73000, 80000, 83545, 95653]
```

```python
plt.figure(figsize=(10, 5))
plt.plot(age, salary_1, "g:", label = "Salary_1")
plt.plot(age, salary_2, "m-.", label = "Salary_2")
plt.xlabel("Age")
plt.xlabel("Slary")
plt.title("SALARY BY AGE") # Başlık
plt.legend(loc="lower right") # legent konumu 
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(age, salary_1, "g:", label = "Salary_1")
plt.plot(age, salary_2, "m-.", label = "Salary_2")
plt.xlabel("Age")
plt.xlabel("Slary")
plt.title("SALARY BY AGE") # Başlık
plt.legend(bbox_to_anchor = (0.45,0.1,0.7,0.1)) # legent konumu ince ayarlar ile yaptık. son iki değer grafik dışı.
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(age, salary_1, "b-", linewidth=6, label = "Salary_1")
plt.plot(age, salary_2, "r-", lw=10, label = "Salary_2")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("SALARY BY AGE")
plt.legend(bbox_to_anchor=(0.45, 0.1, 0.7, 0.1)) #İnce ayar yapmak için
plt.show()
```

# Figure ve Axes Renklendirmesi

```python
plt.figure(facecolor = "yellow", figsize = (8, 4))
ax = plt.axes()
ax.set_facecolor("xkcd:mint green")
plt.plot(age, salary_1, "b-", linewidth=6, label = "Salary_1")
plt.plot(age, salary_2, "r-", lw=3, label = "Salary_2")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("SALARY BY AGE")
plt.legend(loc = "upper center") #İnce ayar yapmak için
plt.show()
```

# Functional Method and Object Oriented (OOP) Methods

### Functional Method

```python
plt.plot(age, salary_1)
plt.xlabel("Age")
plt.ylabel("salary")
plt.title("Salary by Age")
plt.show()
```

### OOP Method

```python
fig, ax = plt.subplots()
ax.plot(age, salary_1)
ax.set_xlabel("Age")
ax.set_ylabel("Salary")
ax.set_title("Salary by Age")
plt.show()
```

### Functional Method (Çoklu Grafik Çizme)

```python
plt. subplot(2,1,1)
plt.plot(age, salary_1, "r")
plt. subplot(2,1,2)
plt.plot(age, salary_1, "b")

plt.show()
```

### OOP Method (Çoklu Grafik Çizme)

```python
fig, ax = plt.subplots(2, 1) # Bririnci değer satır ikinci değer sütun.
ax[0].plot(age, salary_1, "r")
ax[1].plot(age, salary_2, "b")
plt.show()
```

```python
fig, ax = plt.subplots(1, 2)
ax[0].plot(age, salary_1, "r")
ax[1].plot(age, salary_2, "b")
plt.show()
```

```python
fig, ax = plt.subplots(2, 1, figsize=(10,5), facecolor= "yellow")

ax[0].plot(age, salary_1, "r")
ax[1].plot(age, salary_2, "b")


plt.show()
```

 ### OOP Method (Çoklu Grafik Çizme) 2

```python
plt.subplot(2, 3, 1)
plt.plot(age, salary_1, "r")
plt.subplot(2, 3, 2)
plt.plot(age, salary_1, "r")
plt.subplot(2, 3, 3)
plt.plot(age, salary_1, "r")
plt.subplot(2, 3, 4)
plt.plot(age, salary_1, "b")
plt.subplot(2, 3, 5)
plt.plot(age, salary_1, "b")
plt.subplot(2, 3, 6)
plt.plot(age, salary_1, "b")
plt.tight_layout() # içiçe giren kısımları otomatik düzneler
plt.show()
```

```python
fig, ax = plt.subplots(2,3) 
ax[0][0].plot(age,salary_1, "r")
ax[0][1].plot(age,salary_1, "r")
ax[0][2].plot(age,salary_1, "r")

ax[1][0].plot(age,salary_1, "r")
ax[1][1].plot(age,salary_1, "r")
ax[1][2].plot(age,salary_1, "r")


plt.tight_layout() # içiçe giren kısımları otomatik düzneler
plt.show()
```

```python
fig, ax = plt.subplots(2, 3, figsize = (10, 10), facecolor = "yellow")
ax[0][0].plot(age, salary_1, "r")
ax[0][1].plot(age, salary_1, "r")
ax[0][2].plot(age, salary_1, "r")
ax[1][0].plot(age, salary_2, "b")
ax[1][1].plot(age, salary_2, "b")
ax[1][2].plot(age, salary_2, "b")
plt.tight_layout() # içiçe giren kısımları otomatik düzenler
plt.show()
```

```python
x = np.arange(0, 10)
fig, ax = plt.subplots(2, 2, figsize=(12, 6))
ax[0, 0].plot(x, "r", marker = "+", markersize = "5")


ax[0, 1].plot(x-2, "y", marker = "o", markersize = "3")


ax[1, 0].plot(x-3, "blue", marker = "s", markersize = "5")


ax[1, 1].plot(x-4, "g", marker = "1", ms = "5")


plt.show()
```

```python
x = np.arange(0, 30, 0.2)
fig, ax = plt.subplots(figsize=(14, 8))
ax.plot(x, x-1, color = "r", linewidth = 1, linestyle = "-")
ax.plot(x, x-2, color = "green", lw = 2, ls = "-.")
ax.plot(x, x-3, color = "cyan", lw = 3, ls = ":")
ax.plot(x, x-4, color = "m", lw = 4, ls = "--")
plt.show()
```
